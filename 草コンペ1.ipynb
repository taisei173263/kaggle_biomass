```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# „Éë„Çπ„ÅÆË®≠ÂÆö
DATA_DIR = '/content/drive/MyDrive/kaggle/Grass'
train_df = pd.read_csv(os.path.join(DATA_DIR, 'train.csv'))

# 1. „Çø„Éº„Ç≤„ÉÉ„Éà„ÅÆÁ®ÆÈ°û„Åî„Å®„Å´„Éá„Éº„Çø„ÇíÂàÜÈõ¢„Åó„Å¶Áµ±Ë®àÈáè„ÇíÁ¢∫Ë™ç
targets = ['Dry_Green_g', 'Dry_Dead_g', 'Dry_Clover_g', 'GDM_g', 'Dry_Total_g']

print("--- „Çø„Éº„Ç≤„ÉÉ„ÉàÂ§âÊï∞„ÅÆÊ¨†ÊêçÂÄ§„Å®Âü∫Êú¨Áµ±Ë®àÈáè ---")
# Á∏¶ÊåÅ„Å°„Éá„Éº„Çø„ÇíÊ®™ÊåÅ„Å°(Pivot)„Å´Â§âÊèõ„Åó„Å¶Áõ∏Èñ¢„ÇíË¶ã„ÇÑ„Åô„Åè„Åô„Çã
pivot_df = train_df.pivot(index='sample_id', columns='target_name', values='target').reset_index()
print(pivot_df[targets].describe())

# 2. „Çø„Éº„Ç≤„ÉÉ„Éà„ÅÆÂàÜÂ∏É„ÇíÁ¢∫Ë™çÔºàÂèØË¶ñÂåñÔºâ
plt.figure(figsize=(15, 10))
for i, target in enumerate(targets):
    plt.subplot(2, 3, i+1)
    sns.histplot(pivot_df[target].dropna(), kde=True)
    plt.title(f'Distribution of {target}')
plt.tight_layout()
plt.show()
```

- Dry_Clover_g „ÅØÈÄöÂ∏∏„ÅÆÂõûÂ∏∞„É¢„Éá„É´„Åß„ÅØ‰∫àÊ∏¨„ÅåÈõ£„Åó„ÅÑ„Åß„Åô„ÄÇ„Äå„ÇØ„É≠„Éº„Éê„Éº„Åå„ÅÇ„Çã„Åã/„Å™„ÅÑ„Åã„Äç„ÇíÂàÜÈ°û„Åó„Å¶„Åã„ÇâÈáè„Çí‰∫àÊ∏¨„Åô„Çã2ÊÆµÈöé„É¢„Éá„É´„ÇÑ„ÄÅÂ§ñ„ÇåÂÄ§„Å´Âº∑„ÅÑÊêçÂ§±Èñ¢Êï∞ÔºàHuber Loss„Å™„Å©Ôºâ„ÅÆÊ§úË®é„ÅåÂøÖË¶Å„Åã„Å™„Å®


 Gemini„ÅÆÂá∫Âäõ

 „Éê„Ç§„Ç™„Éû„Çπ„ÅÆÊßãÊàê„É´„Éº„É´„ÅÆÁô∫Ë¶ãÔºàÈÄÜÁÆóÔºâÂπ≥ÂùáÂÄ§ÔºàmeanÔºâ„Çí‰Ωø„Å£„Å¶„ÄÅ„Çø„Éº„Ç≤„ÉÉ„ÉàÂ§âÊï∞Èñì„ÅÆÁâ©ÁêÜÁöÑ„Å™Èñ¢‰øÇÂºèÔºà„Éâ„É°„Ç§„É≥„É´„Éº„É´Ôºâ„ÇíÊé®Ê∏¨„Åó„Åæ„Åô„ÄÇ

 Dry_Green_g (26.62) + Dry_Clover_g (6.65) $\approx$ 33.27

 üëâ ‰ªÆË™¨1: $GDM = Green + Clover$ ÔºàÁîü„Åç„ÅüÊ§çÁâ©„ÅÆÂêàË®àÔºâ

 GDM_g (33.27) + Dry_Dead_g (12.04) $\approx$ 45.31

 Dry_Total_g „ÅÆÂπ≥Âùá $\approx$ 45.31

 üëâ ‰ªÆË™¨2: $Total = GDM + Dead$ ÔºàÂÖ®‰Ωì„ÅÆÂêàË®àÔºâ


 ÁµêË´ñ: ÂÖ®„Å¶„ÇíÁã¨Á´ã„Åó„Å¶‰∫àÊ∏¨„Åô„Çã„ÅÆ„Åß„ÅØ„Å™„Åè„ÄÅ**„ÄåÊßãÊàêË¶ÅÁ¥†„Çí‰∫àÊ∏¨„Åó„Å¶Ë∂≥„ÅóÂêà„Çè„Åõ„Çã„Äç„Åæ„Åü„ÅØ„ÄåÂêàË®à„Çí‰∫àÊ∏¨„Åó„Å¶ÊØîÁéá„ÅßÂàÜÈÖç„Åô„Çã„Äç**„Ç¢„Éó„É≠„Éº„ÉÅ„Åå„ÄÅÊï∞Â≠¶ÁöÑÊï¥ÂêàÊÄß„Çí‰øù„Å§„Åü„ÇÅ„ÅÆÈçµ„Åß„Åô„ÄÇ


 üõ°Ô∏è „É°„ÉÄ„É´Áç≤Âæó„Å∏„ÅÆÊà¶Áï•ÁöÑÁµêË´ñ
„Éá„Éº„ÇøÊã°Âºµ„ÅåÂøÖÈ†à: N=357„Åß„ÅØCNN„ÅØ„Åô„Åê„Å´Ë®òÊÜ∂ÔºàÈÅéÂ≠¶ÁøíÔºâ„Åó„Å¶„Åó„Åæ„ÅÑ„Åæ„Åô„ÄÇÁîªÂÉè„ÇíÂõûËª¢„ÄÅÂèçËª¢„ÄÅÊòé„Çã„ÅïÂ§âÊõ¥„Å™„Å©„ÅßÊì¨‰ººÁöÑ„Å´10ÂÄç„Äú100ÂÄç„Å´Â¢ó„ÇÑ„ÅôÂøÖË¶Å„Åå„ÅÇ„Çä„Åæ„Åô„ÄÇ

„Éù„Çπ„Éà„Éó„É≠„Çª„ÉÉ„Ç∑„É≥„Ç∞: ‰∫àÊ∏¨ÂÄ§„ÅåË≤†„ÅÆÂÄ§„Å´„Å™„Å£„ÅüÂ†¥Âêà„ÄÅÂº∑Âà∂ÁöÑ„Å´0„Å´„Åô„ÇãÂá¶ÁêÜ„ÇÑ„ÄÅ‰∏äË®ò„ÅÆ„ÄåÂêàË®à„É´„Éº„É´„Äç„ÇíÊ∫Ä„Åü„Åô„Çà„ÅÜ„Å´ÂÄ§„ÇíË£úÊ≠£„Åô„ÇãÂá¶ÁêÜ„Åå„Çπ„Ç≥„Ç¢„ÇíÊäº„Åó‰∏ä„Åí„Åæ„Åô„ÄÇ

Cross Validation (CV): „Éá„Éº„Çø„ÅåÂ∞ë„Å™„ÅÑ„Åü„ÇÅ„ÄÅ„ÉÜ„Çπ„Éà„Éá„Éº„Çø„Å∏„ÅÆÈÅã„Å´„Çà„Çã„Éñ„É¨„ÅåÂ§ß„Åç„Åè„Å™„Çä„Åæ„Åô„ÄÇ5-FoldÔºà5ÂàÜÂâ≤Ôºâ‰ª•‰∏ä„ÅÆ‰∫§Â∑ÆÊ§úË®º„ÇíË°å„ÅÑ„ÄÅÊâãÂÖÉ„ÅÆ„Çπ„Ç≥„Ç¢ÔºàCV„Çπ„Ç≥„Ç¢Ôºâ„Çí‰ø°„Åò„Çã„Åì„Å®„Åå„ÄÅPrivate LeaderboardÔºàÊúÄÁµÇÈ†Ü‰ΩçÔºâ„Åß„ÅÆËª¢ËêΩ„ÇíÈò≤„Åé„Åæ„Åô„ÄÇ

```python
# --- 1. ‰øÆÊ≠£Áâà„Éá„Éº„Çø„ÅÆ‰ΩúÊàê (image_path„Çí„Ç≠„Éº„Å´„Åô„Çã) ---
# sample_id„Åß„ÅØ„Å™„Åè„ÄÅimage_path„Çí„Ç§„É≥„Éá„ÉÉ„ÇØ„Çπ„Å´„Åó„Å¶Pivot„Åó„Åæ„Åô
pivot_df = train_df.pivot(index='image_path', columns='target_name', values='target').reset_index()

# „Çø„Éº„Ç≤„ÉÉ„ÉàÂêç„É™„Çπ„Éà
targets = ['Dry_Green_g', 'Dry_Dead_g', 'Dry_Clover_g', 'GDM_g', 'Dry_Total_g']

# --- 2. ÊßãÊàê„É´„Éº„É´„ÅÆÂÜçÊ§úË®º ---
# ÁêÜË´ñÂÄ§„ÅÆË®àÁÆó
pivot_df['calc_GDM'] = pivot_df['Dry_Green_g'] + pivot_df['Dry_Clover_g']
pivot_df['calc_Total'] = pivot_df['GDM_g'] + pivot_df['Dry_Dead_g']

# Ë™§Â∑Æ„ÅÆË®àÁÆó
pivot_df['diff_GDM'] = pivot_df['GDM_g'] - pivot_df['calc_GDM']
pivot_df['diff_Total'] = pivot_df['Dry_Total_g'] - pivot_df['calc_Total']

print("--- „Çø„Éº„Ç≤„ÉÉ„ÉàÈñì„ÅÆÊï¥ÂêàÊÄß„ÉÅ„Çß„ÉÉ„ÇØ (‰øÆÊ≠£Áâà) ---")
# „Åª„Å®„Çì„Å©0„Å´Ëøë„ÅÑÂÄ§„ÅåÂá∫„Çã„ÅØ„Åö„Åß„Åô
print(f"GDM - (Green + Clover) Âπ≥ÂùáË™§Â∑Æ: {pivot_df['diff_GDM'].abs().mean():.6f}")
print(f"Total - (GDM + Dead)   Âπ≥ÂùáË™§Â∑Æ: {pivot_df['diff_Total'].abs().mean():.6f}")

# --- 3. Áõ∏Èñ¢Ë°åÂàó„ÅÆÂÜçÁ¢∫Ë™ç ---
# Ê≠£„Åó„ÅèÊï¥ÂΩ¢„Åï„Çå„Åü„Éá„Éº„Çø„ÅßÁõ∏Èñ¢„ÇíË¶ã„Åæ„Åô
plt.figure(figsize=(8, 6))
sns.heatmap(pivot_df[targets].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix (Corrected)")
plt.show()

# --- 4. ÁîªÂÉè„ÅÆÂèØË¶ñÂåñ (ÂÜçÂÆüË°å) ---
# image_path„Åå„Åù„ÅÆ„Åæ„ÅæÂÖ•„Å£„Å¶„ÅÑ„Çã„ÅÆ„Åß„Éû„Éº„Ç∏‰∏çË¶Å
viz_df = pivot_df.copy()
# ÂèØË¶ñÂåñÁî®„ÅÆsample_id„ÅØtrain_df„Åã„ÇâÈÅ©ÂΩì„Å™‰∏Ä„Å§„ÇíÂèñ„Å£„Å¶„Åè„ÇãÔºàË°®Á§∫Áî®Ôºâ
# image_path„Çí„Ç≠„Éº„Å´„Åó„Å¶sample_id„ÇíÂèñÂæó
id_map = train_df[['image_path', 'sample_id']].drop_duplicates(subset='image_path')
viz_df = pd.merge(viz_df, id_map, on='image_path', how='left')

from PIL import Image

def show_images_corrected(df, sort_col, title, n=3):
    top_samples = df.sort_values(by=sort_col, ascending=False).head(n)
    plt.figure(figsize=(15, 5))
    for i, (_, row) in enumerate(top_samples.iterrows()):
        img_name = row['image_path'].split('/')[-1]
        img_path = os.path.join(DATA_DIR, 'train', img_name)
        try:
            img = Image.open(img_path)
            plt.subplot(1, n, i+1)
            plt.imshow(img)
            # ÂÄ§„ÇíË°®Á§∫
            val = row[sort_col]
            plt.title(f"{sort_col}: {val:.1f}g\nTotal: {row['Dry_Total_g']:.1f}g")
            plt.axis('off')
        except Exception as e:
            print(f"Error: {e}")
    plt.suptitle(title)
    plt.show()

print("\n--- ÁîªÂÉè„ÅÆÂèØË¶ñÂåñ (‰øÆÊ≠£Áâà) ---")
show_images_corrected(viz_df, 'Dry_Total_g', 'Top 3 High Biomass Images')
show_images_corrected(viz_df, 'Dry_Clover_g', 'Top 3 High Clover Images')
```
